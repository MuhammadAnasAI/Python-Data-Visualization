```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
titanic = pd.read_csv('titanic.csv')

# Select the numeric features
numeric_features = ['Age', 'Fare', 'Pclass', 'SibSp', 'Parch']

# Create the figure and axes
fig, axes = plt.subplots(len(numeric_features), 2, figsize=(12, 12))

# Create the stem plots
for i, feature in enumerate(numeric_features):
    # Survived
    ax = axes[i, 0]
    sns.scatterplot(x=titanic.loc[titanic['Survived'] == 1, feature], y=[0] * len(titanic.loc[titanic['Survived'] == 1, feature]),
                   marker="|", color="blue", ax=ax)
    ax.set_title(f'{feature} (Survived)')
    ax.set_xlabel('')

    # Not Survived
    ax = axes[i, 1]
    sns.scatterplot(x=titanic.loc[titanic['Survived'] == 0, feature], y=[0] * len(titanic.loc[titanic['Survived'] == 0, feature]),
                   marker="|", color="red", ax=ax)
    ax.set_title(f'{feature} (Not Survived)')
    ax.set_xlabel('')

# Set the y-axis labels
for ax in axes[:,0]:
    ax.set_ylabel(ax.get_title())

# Adjust the spacing between subplots
plt.subplots_adjust(wspace=0.5, hspace=0.5)

# Set the overall title
plt.suptitle('Stem Plots of Titanic Dataset')

# Show the plot
plt.show()
```
