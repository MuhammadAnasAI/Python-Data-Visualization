## Plotting with Matplotlib

```python
# Scatter Plot:
df.plot(kind='scatter', x='Temperature', y='Visibility (km)', color='red')
#Labels
plt.xlabel('Temperature')
plt.ylabel('Visibility')
plt.title('A scatter Plot')
#Save the plot
plt.savefig('./plots/scatterm.png', dpi=300)
#show the plot
plt.show()
```

```python
# box Plot:
df.plot(kind='box', x='Temperature', y='Visibility (km)', color='red', style='Season')
#Labels
plt.xlabel('Temperature')
plt.ylabel('Visibility')
plt.title('A box Plot')
#Save the plot
plt.savefig('./plots/boxm.png', dpi=300)
#show the plot
plt.show()
```

```python
df.columns
```

```python
# Stem Plot with matplotlib:
df.plot(kind='area', x='Temperature', y='Wind Speed', color='red')
#Labels:
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('A kind of Area Plot')
# Save the plot:
plt.savefig('./Plots/AreaPlot.png')
# Show plot:
plt.show()
```

```python
# Stack plot:
df.plot(kind='hist', x='Temperature', y='Humidity', style='Season')
#Lables:
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title('A Histogram Plot')
#Save Plot:
plt.savefig('./Plots/histplot.png')
#Show the Plot:
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Create the 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Temperature'] ,df['Humidity'], df['Wind Speed'])

# Customize the plot
ax.set_xlabel('Temperature')
ax.set_ylabel('Humidity')
ax.set_zlabel('Wind Speed')
ax.set_title('3D Scatter Plot')
# Save the plot:
plt.savefig('./Plots/3dimage.png')

plt.show()
```

```python
# Create a meshgrid of the independent variables
X, Y = np.meshgrid(df['Temperature'], df['Humidity'])

# Convert the 'Wind Speed' column to a 2D array
Z = np.expand_dims(df['Wind Speed'].values, axis=1)

# Create the 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Customize the plot
ax.set_xlabel('Temperature')
ax.set_ylabel('Humidity')
ax.set_zlabel('Wind Speed')
ax.set_title('3D Surface Plot')
# Save the Plot:
plt.savefig('./Plots/surface.png')

plt.show()
```
